# Load Data

```python
# Load data

# Pre, post, and satisfaction assesment Oct 2023
pre = pd.read_csv("../data/Med student pre assessment 2.8.23.csv")
post = pd.read_csv("../data/Advanced post assessment 2.8.23.csv")
satisfaction = pd.read_csv("../data/module satisfaction evaluations 2.8.23.csv")

# Pre and post October 2022 
pre22 = pd.read_csv("../data/IDD-MH Med Student Curriculum Consent_Pre Assessment_October 10, 2022_14.35.csv")
post22 = pd.read_csv("../data/IDD-MH Med Student Advanced Post Assessment_October 10, 2022_14.41.csv")
```

```python
### Pre DATA 
# All questions
preQs = pre.loc[:, pre.columns[pre.columns.str.contains('Q|SC0')]]

# List of questions we have the answer key for that are multiple choice
answerKeyMC = ['Q6', 'Q31', 'Q34', 'Q36', 'Q37', 'Q39', 'Q12', 'Q13', 'Q22', 'Q23']
preQsMC = preQs.loc[:, preQs.columns.isin(answerKeyMC)]

# Read in Answer Key CSV
preAnswerKey = pd.read_csv('../data/preAssessAnswerKey.csv').head()

for question in preQsMC.columns:
    if question in preAnswerKey.columns:
        new_column_name = f"{question}_correct"
        preQsMC[new_column_name] = np.where(preQsMC[question].isin(preAnswerKey[question]), 1, 0)
        correct_answers = '|'.join(str(ans) for ans in preAnswerKey[question].dropna())

        
# drop NA rows and add score column     
preQsMC = preQsMC.dropna(subset=['Q6'])
preQsMC['score'] = preQsMC.filter(like = "correct").sum(axis = 1)
clean_pre_qs_scored = preQsMC.drop([0, 1], axis=0).reset_index(drop=True)

# table with columns of how people did on questions
pre_scores = clean_pre_qs_scored.loc[:, clean_pre_qs_scored.columns[clean_pre_qs_scored.columns.str.contains('correct')]].mean()
pre_scores = pre_scores * 100
pre_scores
```

```python
# POST DATA CLEANING AND SCORING
# All Questions
postQs = post.loc[:, post.columns[post.columns.str.contains('Q|SC0')]]

# Rename columns that have the same name
postQs.rename(columns = {'Q56' : 'Q56_1'}, inplace = True)
postQs.rename(columns = {'Q56.1' : 'Q56_2'}, inplace = True)
postQs.rename(columns = {'Q60' : 'ZZZ'}, inplace = True)
postQs.rename(columns = {'Q36' : 'Q36X'}, inplace = True)
postQs.rename(columns = {'Q31' : 'Q31X'}, inplace = True)
postQs.rename(columns = {'Q37' : 'Q37X'}, inplace = True)

# Dictionary to match question names with 
rename_dict = {
    'Q56_2': 'Q6',
    'Q56_1': 'Q34',
    'Q42': 'Q31',
    'ZZZ': 'Q36',
    'Q58': 'Q37',
    'Q62': 'Q39',
    'Q50': 'Q12',
    'Q52': 'Q13',
    'Q64': 'Q22',
    'Q66': 'Q23'
}

# Subset to only answer key questions
postQs.rename(columns=rename_dict, inplace = True)
subRenamePostQs = postQs.loc[:, postQs.columns.isin(answerKeyMC)]

# Read in Answer Key CSV
postAnswerKey = pd.read_csv('../data/postAssessAnswerKey.csv').head()

# Add column for correct answer or not
for question in subRenamePostQs.columns:
    if question in postAnswerKey.columns:
        new_column_name = f"{question}_correct"
        subRenamePostQs[new_column_name] = np.where(subRenamePostQs[question].isin(postAnswerKey[question]), 1, 0)
        correct_answers = '|'.join(str(ans) for ans in postAnswerKey[question].dropna())

#         Clean and add score columns 
subRenamePostQs = subRenamePostQs.dropna(subset=['Q6'])
subRenamePostQs['score'] = subRenamePostQs.filter(like = "correct").sum(axis = 1)
clean_post_qs_scored = subRenamePostQs.drop([0, 1, 2], axis=0).reset_index(drop=True)

# table with columns of how people did on questions
post_scores = clean_post_qs_scored.loc[:, clean_post_qs_scored.columns[clean_post_qs_scored.columns.str.contains('correct')]].mean()
post_scores = post_scores * 100
post_scores
```

```python
import matplotlib.pyplot as plt

# Order quetions
custom_order = ['Q6_correct', 'Q12_correct', 'Q13_correct', 'Q22_correct', 'Q23_correct', 
                'Q31_correct', 'Q34_correct', 'Q36_correct', 'Q37_correct', 'Q39_correct']
custom_labels = ['Q6', 'Q12', 'Q13', 'Q22', 'Q23', 'Q31', 'Q34', 'Q36', 'Q37', 'Q39']
# plot
fig, ax = plt.subplots()
bar_width = 0.35
index = range(len(pre_scores))



bar1 = ax.bar(index, pre_scores[custom_order], bar_width, label='Pre-Scores', color = "red")  # Use the Series values directly
bar2 = ax.bar(index, post_scores[custom_order], bar_width, label='Post-Scores', alpha=.7, align='edge', color = "blue")  # Adjust alignment and alpha for visibility

# Customize the plot
ax.set_xlabel('Questions')
ax.set_ylabel('Percentage (%)')
ax.set_title('Difference between Pre and Post Assessment Questions')
ax.set_xticks(index)
ax.set_xticklabels(custom_labels, rotation=45, ha='right')  # Use the Series index for x-axis labels
ax.legend()
```

```python
# Module Satisfaction Evaluations 
sat = satisfaction
sat['satisfaction'] = sat['Q2']

satisfaction_mapping = {
    'Not Satisfied': 1,
    'Somewhat Satisfied': 2,
    'Satisfied': 3,
    'Highly satisfied': 4
}

sat['satisfaction'] = sat['satisfaction'].map(satisfaction_mapping)

plt.figure(figsize=(12, 8))
sns.barplot(x='satisfaction', y='Q1', data=sat, ci=None)

# Add labels and title
plt.xlabel('Satisfaction Level')
plt.ylabel('Training Module')
plt.title('Satisfaction Levels for Each Training Module')


satisfaction['Q1'].value_counts()
```
